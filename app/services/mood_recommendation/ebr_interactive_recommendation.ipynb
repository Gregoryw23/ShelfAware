```python
def interactive_recommendation():
    """
    Interactive function for users to get recommendations
    """
    print("\n" + "="*70)
    print("ðŸ“š MOOD-BASED BOOK RECOMMENDATION SYSTEM")
    print("="*70)

    print("\nHow would you like to input your mood?")
    print("1. Describe your mood in words")
    print("2. Select from emotion categories")

    choice = input("\nYour choice (1 or 2): ")

    if choice == '1':
        mood_text = input("\nDescribe how you're feeling: ")
        mode_choice = input("Do you want books that (1) match or (2) contrast your mood? ")
        mode = 'match' if mode_choice == '1' else 'contrast'

        recommendations = recommender.recommend_by_text(mood_text, top_n=5, mode=mode)
        recommender.display_recommendations(recommendations, mode=mode)

    elif choice == '2':
        print("\nAvailable emotions:")
        emotions = list(emotion_lexicon.keys())
        for i, emotion in enumerate(emotions, 1):
            print(f"{i}. {emotion}", end="  ")
            if i % 5 == 0:
                print()

        print("\n\nEnter emotion numbers (comma-separated) that describe your mood:")
        selected = input("Your selection: ")

        user_mood = {}
        for idx in selected.split(','):
            try:
                emotion_idx = int(idx.strip()) - 1
                if 0 <= emotion_idx < len(emotions):
                    user_mood[emotions[emotion_idx]] = 100
            except:
                pass

        mode_choice = input("\nDo you want books that (1) match or (2) contrast your mood? ")
        mode = 'match' if mode_choice == '1' else 'contrast'

        recommendations = recommender.recommend_by_mood(user_mood, top_n=5, mode=mode)
        recommender.display_recommendations(recommendations, mode=mode)

# Run interactive recommendation
# Uncomment the line below to run interactively
# interactive_recommendation()
```
